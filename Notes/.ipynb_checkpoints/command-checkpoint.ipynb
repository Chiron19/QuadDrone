# Simulation Pace

自己添加模拟速率模块，以便观察模拟曲线。

# Pilot Joystick

连接控制手柄（以ThrustMaster A320为例），取出roll，pitch，yaw（做积分）和throttle并汇到总线。

在结构体command中调节变量以控制死区、增益。

# Bus:AC Cmd

## controlModePosVSOrient

default:0

## pos_ref

基准位置

default:"positionRef.mat"

## takeoff_flag

起飞标记，起飞时标记为1，调节变量takeOffDuration以调节起飞时间

default:(time < takeOffDuration)?

## orient_ref

default:[yaw,pitch,roll]

## live_time_ticks

定时清零计时器（32 bits）
